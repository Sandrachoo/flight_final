## Add in Day of Week on Departure Date

```python
# Function to get the day of the week from the departure time
def get_day_of_week(departure_time):
    date_obj = datetime.strptime(departure_time, '%Y-%m-%d %H:%M')
    return date_obj.strftime('%A')
```

```python
import pandas as pd
from datetime import datetime

# List of CSV files to process
csv_files = [
    'flights_KUL_to_BKK1.csv', 'flights_KUL_to_BKK2.csv',
    'flights_KUL_to_CGK1.csv', 'flights_KUL_to_CGK2.csv',
    'flights_KUL_to_BWN1.csv', 'flights_KUL_to_BWN2.csv',
    'flights_KUL_to_MNL1.csv', 'flights_KUL_to_MNL2.csv',
    'flights_KUL_to_PNH1.csv', 'flights_KUL_to_PNH2.csv',
    'flights_KUL_to_RGN1.csv', 'flights_KUL_to_RGN2.csv',
    'flights_KUL_to_SGN1.csv', 'flights_KUL_to_SGN2.csv',
    'flights_KUL_to_SIN1.csv', 'flights_KUL_to_SIN2.csv',
    'flights_KUL_to_VTE1.csv', 'flights_KUL_to_VTE2.csv',
]

# Function to convert datetime format
def convert_datetime_format(date_str):
    try:
        # Convert from current format to the desired format
        return datetime.strptime(date_str, '%d/%m/%Y %H:%M').strftime('%Y-%m-%d %H:%M')
    except ValueError:
        return date_str  # If conversion fails, return the original string

# Loop through each CSV file
for file in csv_files:
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Convert Departure Time and Arrival Time formats
    df['Departure Time'] = df['Departure Time'].apply(convert_datetime_format)
    df['Arrival Time'] = df['Arrival Time'].apply(convert_datetime_format)
    
    # Save the updated DataFrame back to the CSV file
    df.to_csv(file, index=False)

print("Datetime format conversion completed for all files.")
```

```python
# Apply the function to the 'Departure Time' column and create a new column for the day of the week
singapore_data1['Day of week (on flight date)'] = singapore_data1['Departure Time'].apply(get_day_of_week)

# Save the updated DataFrame back to the CSV
singapore_data1.to_csv('flights_KUL_to_SIN1_updated.csv', index=False)


print("Updated CSV with day of the week saved as flights_KUL_to_SIN1_updated.csv")
```

## Add in If Departure Date is a Holiday

```python
# Function to check if a date is a holiday based on the given list of holidays
def is_holiday(date_str, holiday_list):
    date_obj = datetime.strptime(date_str, '%Y-%m-%d %H:%M')
    return date_obj.strftime('%Y-%m-%d') in holiday_list
```

```python
import pandas as pd
from datetime import datetime

# Sample lists of national holidays in Malaysia
malaysia_departure_country_holidays = [
    '2024-07-07',  # Awal Muharram
    '2024-07-08',  # Awal Muharram Holiday
    '2024-08-31',  # National Day 
]

# Sample lists of holidays in Singapore
singapore_arrival_country_holidays = [
    '2024-08-09'  # National Day 
]
```

```python
# Load the CSV file into a DataFrame
singapore_data1 = pd.read_csv('flights_KUL_to_SIN1_updated.csv')

# Apply the function to the 'Departure Time' and 'Arrival Time' columns using the respective holiday lists
singapore_data1['Departure Country Is Holiday'] = singapore_data1['Departure Time'].apply(lambda x: is_holiday(x, malaysia_departure_country_holidays))
singapore_data1['Arrival Country Is Holiday'] = singapore_data1['Arrival Time'].apply(lambda x: is_holiday(x, singapore_arrival_country_holidays))

# Save the updated DataFrame back to the CSVd
singapore_data1.to_csv('flights_KUL_to_SIN1_updated.csv', index=False)

print("Updated CSV with holiday checks saved as flights_KUL_to_SIN1_updated.csv")
```

```python
# Function to check if a flight is overnight
def is_overnight(departure_time, arrival_time):
    try:
        # Attempt to parse using the format '%Y-%m-%d %H:%M:%S'
        departure_datetime = datetime.strptime(departure_time, '%Y-%m-%d %H:%M:%S')
        arrival_datetime = datetime.strptime(arrival_time, '%Y-%m-%d %H:%M:%S')
    except ValueError:
        try:
            # Attempt to parse using the format '%Y-%m-%d %H:%M'
            departure_datetime = datetime.strptime(departure_time, '%Y-%m-%d %H:%M')
            arrival_datetime = datetime.strptime(arrival_time, '%Y-%m-%d %H:%M')
        except ValueError:
            raise ValueError(f"Date format not recognized for departure: {departure_time} or arrival: {arrival_time}")

    # Check if the arrival date is the day after the departure date
    return arrival_datetime.date() > departure_datetime.date()
```

```python
import pandas as pd
from datetime import datetime

# Load the CSV file into a DataFrame
singapore_data1 = pd.read_csv('flights_KUL_to_SIN1_updated.csv')

# Apply the function to the DataFrame to create the 'Overnight Flight' column
singapore_data1['Overnight Flight'] = singapore_data1.apply(lambda row: is_overnight(row['Departure Time'], row['Arrival Time']), axis=1)

# Save the updated DataFrame back to the CSV file
singapore_data1.to_csv('flights_KUL_to_SIN1_updated.csv', index=False)

print("Updated CSV with 'Overnight Flight' column added saved as flights_KUL_to_SIN1_updated.csv")
```
